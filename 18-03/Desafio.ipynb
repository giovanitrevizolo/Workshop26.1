```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
temp = []
for x in range (1,8):
  x = random.randint(20,35)
  temp.append(x)
temperatura = np.array(temp)

dia = ["Domingo", "Segunda" , "Terça" , "Quarta" , "Quinta" , "Sexta" , "Sábado"]
```

```python
print(temperatura)
print(dia)
```

```python
data = {
    "Dia" : dia ,
    "Temperatura" : temperatura
}

df = pd.DataFrame(data)

print(df)
```

```python
df.describe()
```

```python
plt.plot(dia,temperatura, marker = 's')
plt.show()
```

```python
plt.bar(dia, temperatura)
plt.show()
```

```python
plt.scatter(dia , temperatura)
plt.grid(color = 'red', axis = 'y')
plt.show()
```
